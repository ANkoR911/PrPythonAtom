```python
class HashMap:

    class Entry:
        
        def __init__(self, key, value):
            self.key = key
            self.value = value

        def get_key(self):
            return self.key

        def get_value(self):
            return self.value

        def __eq__(self, other):
            return (self.key == other.key)

    def __init__(self, bucket_num = 64):
        self.capacity = bucket_num
        self.base = []
        for i in range(self.capacity):
            self.base.append([])
        self.counter = 0
    

    def testsize(self):
        for i in range(self.capacity):
            print(len(self.base[i]))


    def get(self, key, default_value=None):
        index = self._get_hash(key)
        Test = True
        for i in self.base[index]:
            if i.key == key:
                return i.value
                Test = False
        if Test:
            return default_value
    

    def put(self, key, value):
        index = self._get_hash(key)
        newEntry = self.Entry(key, value)
        Test = True
        for i in self.base[index]:
            if i.key == key:
                i.value = value
                Test = False
        if Test:
            self.counter += 1    
            self.base[index].append(newEntry)

    


    def __len__(self):
        return self.counter
   

    def _get_hash(self, key):
        return (hash(key)%self.capacity)
    

    def _get_index(self, hash_value):
        return hash_value%self.capacity
    

    def generator(self, val):
        start = 0
        while start < val:
            yield start
            start += 1

    def generator_values(self):
        Rez = []
        gen1 = self.generator(self.capacity)
        for i in gen1:
            gen2 = self.generator(len(self.base[i]))
            for j in gen2:
                Rez.append(self.base[i][j].value)
        return Rez

    def generator_keys(self):
        Rez = []
        gen1 = self.generator(self.capacity)
        for i in gen1:
            gen2 = self.generator(len(self.base[i]))
            for j in gen2:
                Rez.append(self.base[i][j].key)
        return Rez

    def generator_items(self):
        Rez = []
        gen1 = self.generator(self.capacity)
        for i in gen1:
            gen2 = self.generator(len(self.base[i]))
            for j in gen2:
                Rez.append((self.base[i][j].key, self.base[i][j].value))
        return Rez

    def values(self):
        return self.generator_values()
   
    def keys(self):
        return self.generator_keys()
    
    def items(self):
        return self.generator_items()
   
    def __str__(self):
        Rez = ''
        for i in self.items():
            Rez += '({}, {})'.format(i[0], i[1])
        return Rez
```

```python
import random

a = HashMap()
#a.testsize()
a.put(2, 5)
#a.testsize()
a.put(3, 13)
a.put(4, 14)
a.put(5, 15)
a.put(6, 16)
#a.testsize()
a.put(2, 7)
#print(a.get(2, 10))
#print(a.get(5, 10))
#print(a.__len__())
#print("Hello")
print(a.values())
print("Hello")
print(a.keys())
print("Hello")
print(a.items())
print(a)

def test_hashmap_04():
    entries = [(5, 7), ("entries", 56), ("value", 54.), (1000, "t"), (HashMap(10), ()), ({"s":"v"}, {"v":"s"})]
    for k, v in entries:
        entry = HashMap.Entry(k, v)
        assert entry.get_key() == k
        assert entry.get_value() == v
    print("Test 4 part 01 passed")
    for i in range(len(entries)):
        entry_one = HashMap.Entry(entries[i][0], entries[i][1])
        for _ in range(10):
            j = random.randint(0, len(entries)-1)
            p = random.randint(0, len(entries)-1)
            entry_two = HashMap.Entry(entries[j][0], entries[p][1])
            if j == i:
                assert entry_one == entry_two
            else:
                assert entry_one != entry_two
    print("Test 4 part 02 passed")

def test_hashmap_05():
    hashmap = HashMap(10)
    assert sum(isinstance(v, list) for k, v in hashmap.__dict__.items()) == 1
    print("Test 5 part 01 passed")
    inner_list_name = [k for k, v in hashmap.__dict__.items() if isinstance(v, list)][0]
    for i in range(10):
        assert len(HashMap(i).__dict__[inner_list_name]) == i
    print("Test 5 part 02 passed")



def test_hashmap_06():
    hashmap = HashMap(10)
    entries = [(5, 7), ("entries", 56), ("value", 54.), (1000, "t"), (HashMap(10), ())]
    for k, v in entries:
        hashmap.put(k, v)
    for k, v in entries:
        assert hashmap.get(k) == v
    print("Test 6 part 01 passed")
    for _ in range(100):
        i = random.randint(0, len(entries)-1)
        j = random.randint(0, len(entries)-1)
        hashmap.put(i, j)
        assert hashmap.get(i) == j
    assert hashmap.get("nexit", "default") =="default"
    print("Test 6 part 02 passed")
```
